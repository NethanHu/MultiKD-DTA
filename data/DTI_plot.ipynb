```python
from scipy.stats import norm
import numpy as np

data = DTI(name = 'Davis')
df = data.get_data()

plt.figure(dpi=300,figsize=(21, 5))

plt.subplot(1, 3, 1)
lengths = df['Drug'].apply(len)
xmin, xmax = 0, 100
plt.hist(lengths, bins=20, color='skyblue', edgecolor='white', density=True, range=(xmin, xmax))
# 
mu, sigma = norm.fit(lengths)
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
# 
plt.plot(x, p, 'k', linewidth=2, color='blue')
plt.xlabel('Length of ligands SMILES')
plt.ylabel('Number of Ligands')
plt.grid(True)


plt.subplot(1, 3, 2)
lengths = df['Target'].apply(len)
xmin, xmax = 0, 3000
plt.hist(lengths, bins=20, color='skyblue', edgecolor='white', density=True, range=(xmin, xmax))
# 
mu, sigma = norm.fit(lengths)
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
# 
plt.plot(x, p, 'k', linewidth=2, color='blue')
plt.xlabel('Length of protein sequences')
plt.ylabel('Number of Proteins')
plt.grid(True)

plt.subplot(1, 3, 3)
lengths = df['Y']
xmin, xmax = -5000, 25000
plt.hist(lengths, bins=20, color='skyblue', edgecolor='white', density=True, range=(xmin, xmax))
# 
mu, sigma = norm.fit(lengths)
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)
# 
plt.plot(x, p, 'k', linewidth=2, color='blue')
plt.xlabel('Affinity')
plt.ylabel('Number of DTA Pairs')
plt.grid(True)
    
plt.show()
```
